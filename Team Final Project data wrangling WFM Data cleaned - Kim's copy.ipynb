### What kind of devices are you using to connect to your corporate working environment? 

#### Determine levels for Devices variable

```python
df.Devices.value_counts()
```

#### Recoding for Devices variable

```python
def Devices_recode (series):
    if series == "Laptop;":
        return 0
    if series == "Laptop;Smartphone;":
        return 1
    if series == "Desktop;":
        return 2
    if series == "Desktop;Laptop;":
        return 3
    if series == "Desktop;Laptop;Tablet;Smartphone;":
        return 4
    if series == "Laptop;Tablet;":
        return 5
    if series == "Laptop;Tablet;Smartphone;":
        return 6
    if series == "Desktop;Laptop;Smartphone;":
        return 7
    if series == "Desktop;Laptop;Tablet;Smartphone;Other;":
        return 8
    if series == "Desktop;Smartphone;":
        return 9
    if series == "Smartphone;":
        return 10
    
df["DevicesR"] = df["Devices"].apply(Devices_recode)
```
